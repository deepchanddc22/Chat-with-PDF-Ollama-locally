```python
import pdfplumber
from llama_index.embeddings.ollama import OllamaEmbedding

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    text = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text.append(page.extract_text())
    return "\n".join(text)

# Initialize OllamaEmbedding
ollama_embedding = OllamaEmbedding(
    model_name="llama3",
    base_url="http://localhost:11434",
    ollama_additional_kwargs={"mirostat": 0},
)

# Extract text from the PDF
pdf_path = "/home/deepchanddc2/projects/chatapp/data/copyrightrules1957.pdf"
pdf_text = extract_text_from_pdf(pdf_path)

# Assuming the PDF content is divided into passages (e.g., by paragraphs)
passages = pdf_text.split('\n\n')  # Splitting by double newline for paragraphs

# Get embeddings for the extracted passages
pass_embedding = ollama_embedding.get_text_embedding_batch(passages, show_progress=True)
print(pass_embedding)

# Get an embedding for a query
query_embedding = ollama_embedding.get_query_embedding("Where is the price of cybertruck?")
print(query_embedding)
```

```python
import faiss
import numpy as np

pass_embeddings = ollama_embedding.get_text_embedding_batch(passages, show_progress=True)

# Convert embeddings to numpy array
pass_embeddings_array = np.array(pass_embeddings)

# Initialize FAISS index
dimension = pass_embeddings_array.shape[1]  # Embedding dimension
index = faiss.IndexFlatL2(dimension)

# Add embeddings to the index
index.add(pass_embeddings_array)

def get_nearest_neighbor(query):
    query_embedding = ollama_embedding.get_query_embedding(query)
    query_embedding_array = np.array([query_embedding])
    distances, indices = index.search(query_embedding_array, 1)  # Retrieve only the closest passage
    closest_passage_index = indices[0][0]
    closest_passage = passages[closest_passage_index]
    distance = distances[0][0]
    return closest_passage, distance

# Example query
query = "Where is the price of cybertruck?"
closest_passage, distance = get_nearest_neighbor(query)

# Print the retrieved passage and its distance
print(f"Closest Passage: {closest_passage}")
print(f"Distance: {distance}")
print("-" * 80)
```

```python
import json
import numpy as np
import ollama

# Documents to be embedded
documents = [
    "Llamas are members of the camelid family meaning they're pretty closely related to vicuñas and camels",
    "Llamas were first domesticated and used as pack animals 4,000 to 5,000 years ago in the Peruvian highlands",
    "Llamas can grow as much as 6 feet tall though the average llama between 5 feet 6 inches and 5 feet 9 inches tall",
    "Llamas weigh between 280 and 450 pounds and can carry 25 to 30 percent of their body weight",
    "Llamas are vegetarians and have very efficient digestive systems",
    "Llamas live to be about 20 years old, though some only live for 15 years and others live to be 30 years old",
]

# Generate embeddings and store them locally
data = []
for i, d in enumerate(documents):
    response = ollama.embeddings(model="mxbai-embed-large", prompt=d)
    embedding = response["embedding"]
    data.append({
        "id": str(i),
        "document": d,
        "embedding": embedding
    })

# Save the data to a JSON file
with open("embeddings.json", "w") as f:
    json.dump(data, f, indent=4)

print("Embeddings and documents have been saved to embeddings.json")

# Load the data from the JSON file
with open("embeddings.json", "r") as f:
    data = json.load(f)

# Function to calculate cosine similarity
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# Function to query the documents
def query_documents(query, top_k=3):
    # Generate embedding for the query
    response = ollama.embeddings(model="nomic-embed-text", prompt=query)
    query_embedding = response["embedding"]

    # Calculate similarity scores
    similarities = []
    for entry in data:
        embedding = np.array(entry["embedding"])
        similarity = cosine_similarity(query_embedding, embedding)
        similarities.append((similarity, entry["document"]))

    # Sort by similarity score
    similarities.sort(reverse=True, key=lambda x: x[0])

    # Return the top_k most similar documents
    return similarities[:top_k]

# Example query
query = "How tall can llamas grow?"
results = query_documents(query)

for score, doc in results:
    print(f"Score: {score:.4f}, Document: {doc}")
```
